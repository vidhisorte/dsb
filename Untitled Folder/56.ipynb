```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
d={
    'a':[10,26,34,58,1,84],
    'b':[12,6,47,35,14,4],
    'c':[5,61,30,24,15,24],
    'd':[5,5,5,10,10,10],
}
```

```python
df=pd.DataFrame(d)
```

```python
df
```

```python
def meann(lst):
    ans=0;
    for i in range(len(lst)):
        ans+=lst[i]
    ans=ans/len(lst)
    return ans

for i in df:
    print(meann(list(df[i])))
```
