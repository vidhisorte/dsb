```python
ut1=df['v'].quantile(0.25)
ut2=df['v'].quantile(0.75)

iqr=ut2-ut1

lb=ut1-1.5*iqr
ub=ut2+1.5*iqr

print(lb,ub)
```

```python
t=df
for i in range(len(t['v'])):
    
    if (t['v'][i]<lb or t['v'][i]>ub):
        print(t['v'][i])
        t.drop(i,inplace=True)
```
